# Show images


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
images_orig = np.load('data/UCMerced_images_orig_MultiClass.npy')
nb_classes = images_orig.shape[3]
num_images = int(images_orig.shape[0]/nb_classes)

def plotImages(images, num_images=5):
    fig, axes = plt.subplots(1, num_images, figsize=(10, 10))
    axes = axes.flatten()
    for img, ax in zip(images, axes):
        ax.imshow(img)
        ax.set_xticks(())
        ax.set_yticks(())
    plt.tight_layout()
    plt.show()

def plotSamplesPerCategory():
    for i in range(nb_classes):
        print("class: ", i)
        images_to_print = images_orig[i*num_images:(i+1)*num_images]
        np.random.seed(2020)
        np.random.shuffle(images_to_print)
        print(images_to_print.shape)
        plotImages(images_to_print)

plotSamplesPerCategory()
```
